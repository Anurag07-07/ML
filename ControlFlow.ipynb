```python
# while loop

l = [1,2,3,4,5]
start = 0;end  = len(l)-1
while start<end:
    sum = l[start]+l[end]
    print(sum)
    start = start +1
    end = end - 1
```

```python
# along with for loop we can use else block 

def prime(n):
    for i in range(2,int(n**(1/2))+1):
        if n%i==0:
            return False
    return True            

for i in range(1,101):
    if prime(i):
        print(i,end=" ")
    else:
        pass
       
```

```python
# Factorial Calculation

n = 5
i = 1
fac = 1;

while i<=n:
    fac = fac * i
    i = i+1

print(fac)
```

```python
# sum of digits
n = int(input("Enter The Number"))

ans = 0
while n!=0:
    last = n%10;
    ans = ans + last
    n = n//10

print(f"The Sum is {ans}")
```
